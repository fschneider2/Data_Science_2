Utilize os arquivos funcionarios.csv, cargos.csv e setores.csv como
entrada para o processamento requerido nos seguintes exercícios.

1) Faça o pivoteamento de forma a mostrar a relação entre cargos e setores
com a média salarial.
2) Faça um pivoteamento para elucidar quais setores possuem mais de 5
funcionários acima de 55 anos?
3) Faça um pivoteamento para mostrar a relação entre os níveis dos cargos e
os salários dos funcionários com a média salarial.

```python
import pandas as pd 
import numpy as np 
```

```python
df_cargos = pd.read_csv('/mnt/2819a183-0023-410d-bc7c-d6fcca623389/Repositorios - Projetos/Data_Science_2/Semana_2/cargos.csv')
df_setores = pd.read_csv('/mnt/2819a183-0023-410d-bc7c-d6fcca623389/Repositorios - Projetos/Data_Science_2/Semana_2/setores.csv')
df_funcionarios = pd.read_csv('/mnt/2819a183-0023-410d-bc7c-d6fcca623389/Repositorios - Projetos/Data_Science_2/Semana_2/funcionarios.csv')
```

```python
display(df_cargos,df_funcionarios,df_setores)
```

Criando o DataFrame completo:

```python
df_funcionarios_completo = df_funcionarios.merge(df_cargos)

df_funcionarios_completo = df_funcionarios_completo.merge(df_setores)

df_funcionarios_completo.head(1)
```

1) Faça o pivoteamento de forma a mostrar a relação entre cargos e setores
com a média salarial.

```python
mean_wage_cargos_setores = pd.pivot_table(df_funcionarios_completo,
    index=['cargo_nome'],
    columns=['setor_nome'],
    # aggfunc={'cargo_salario' : np.mean}
    aggfunc={'cargo_salario' : np.sum}
)
display(mean_wage_cargos_setores)
```

2) Faça um pivoteamento para elucidar quais setores possuem mais de 5
funcionários acima de 55 anos?

```python
df_more_55_years = df_funcionarios_completo.query('funcionario_idade > 55')

department_empl_be_more_55_years = pd.pivot_table(
    df_more_55_years,
    index= ['setor_nome'],
    columns=['cargo_nome'],
    aggfunc = {'funcionario_idade': np.size}
)
display(department_empl_be_more_55_years)
```

3) Faça um pivoteamento para mostrar a relação entre os níveis dos cargos e
os salários dos funcionários com a média salarial.

```python
mean_wage_cargos_nivel = pd.pivot_table(
    df_funcionarios_completo,
        index=['cargo_nivel'],
        columns=['setor_nome'],#para anlise da média dos niveis por setor
        # columns=['cargo_nivel'],#para analise da média exclusiva dos niveis 
        aggfunc={'cargo_salario' : np.mean}
        # aggfunc={'cargo_salario' : np.sum} # para analise do custo total
)
display(mean_wage_cargos_nivel)
```
